```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('./wandb-log.csv')

# Extract the accuracy values from the data
accuracy_label_mixup = data['plot-label-mixup - test_acc']
accuracy_no_uni_pred = data['plot-no-uni-pred - test_acc']
accuracy_100 = data['plot-100% - test_acc']
accuracy_33 = data['plot-33% - test_acc']
accuracy_0 = data['plot-0% - test_acc']
# Filter the data based on the condition step <= 45000
filtered_data = data[data['Step'] <= 45000]

# Create a line plot of the accuracy values
plt.plot(filtered_data['Step'], filtered_data['plot-label-mixup - test_acc'], label='Label Mixup')
plt.plot(filtered_data['Step'], filtered_data['plot-no-uni-pred - test_acc'], label='No Unlabeled Prediction')
plt.plot(filtered_data['Step'], filtered_data['plot-100% - test_acc'], label='100% Unlabeled Data')
plt.plot(filtered_data['Step'], filtered_data['plot-33% - test_acc'], label='33% Unlabeled Data')
plt.plot(filtered_data['Step'], filtered_data['plot-0% - test_acc'], label='0% Unlabeled Data')

# Add labels and title to the plot
plt.xlabel('Step')
plt.ylabel('Accuracy')
plt.title('Accuracy Plot')

# Display the plot
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
plt.style.use('seaborn-paper')
plt.rcParams['font.family'] = 'serif'
plt.figure(figsize=(7, 5))

plt.rcParams['axes.labelpad'] = 10
plt.rcParams['axes.titlesize'] = 15
plt.rcParams['axes.titlepad'] = 10
# Read the CSV file
data = pd.read_csv('./wandb-log.csv')

# Filter the data based on the condition step <= 45000
data = data[data['Step'] <= 45000]

accuracy_100 = data['plot-100% - test_acc']
accuracy_33 = data['plot-33% - test_acc']
accuracy_0 = data['plot-0% - test_acc']

# Calculate the moving average of the accuracy values
window_size = 6
smoothed_accuracy_100 = accuracy_100.ewm(span=window_size).mean()
smoothed_accuracy_33 = accuracy_33.ewm(span=window_size).mean()
smoothed_accuracy_0 = accuracy_0.ewm(span=window_size).mean()

# Create a line plot of the smoothed accuracy values
plt.plot(data['Step'], smoothed_accuracy_100*100, label='100% Unlabeled Data')
plt.plot(data['Step'], smoothed_accuracy_33*100, label='33% Unlabeled Data')
plt.plot(data['Step'], smoothed_accuracy_0*100, label='0% Unlabeled Data')

plt.plot(data['Step'], accuracy_100*100, label='100% Unlabeled Data', alpha=0.05)
plt.plot(data['Step'], accuracy_33*100, label='33% Unlabeled Data', alpha=0.05)
plt.plot(data['Step'], accuracy_0*100, label='0% Unlabeled Data', alpha=0.05)

# Add labels and title to the plot
plt.xlabel('Step')
plt.ylabel('Accuracy')
plt.title('Smoothed Accuracy Plot')

plt.legend(['100% Unlabeled Data', '33% Unlabeled Data', '0% Unlabeled Data'])

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Apply seaborn-paper style and configure plot aesthetics
plt.style.use('seaborn-paper')
plt.rcParams['font.family'] = 'serif'
plt.figure(figsize=(6, 4))

# Configure axis padding and title sizes
plt.rcParams['axes.labelpad'] = 12
plt.rcParams['axes.titlesize'] = 18
plt.rcParams['axes.titlepad'] = 15

# Read the CSV file
data = pd.read_csv('./wandb-log.csv')

# Filter the data based on the condition step <= 45000
data = data[data['Step'] <= 45000]

# Extract accuracy columns
accuracy_100 = data['plot-100% - test_acc']
accuracy_33 = data['plot-33% - test_acc']
accuracy_0 = data['plot-0% - test_acc']

# Calculate the moving average of the accuracy values
window_size = 7
smoothed_accuracy_100 = accuracy_100.ewm(span=window_size).mean()
smoothed_accuracy_33 = accuracy_33.ewm(span=window_size).mean()
smoothed_accuracy_0 = accuracy_0.ewm(span=window_size).mean()

# Create a line plot of the smoothed accuracy values with enhanced visualization
plt.plot(data['Step'], smoothed_accuracy_0 * 100, label='Only MultiSoftClip', color='red', linewidth=2)
plt.plot(data['Step'], smoothed_accuracy_100 * 100, label='Only M3Co', color='blue', linewidth=2)
plt.plot(data['Step'], smoothed_accuracy_33 * 100, label='0.33 M3Co + 0.67 MultiSoftClip', color='green', linewidth=2)

# Add original accuracy values with low opacity for context
plt.plot(data['Step'], accuracy_100 * 100, color='blue', alpha=0.1)
plt.plot(data['Step'], accuracy_33 * 100, color='green', alpha=0.1)
plt.plot(data['Step'], accuracy_0 * 100, color='red', alpha=0.1)

# Add labels and title to the plot
plt.xlabel('Step', fontsize=14)
plt.ylabel('Accuracy (%)', fontsize=14)
plt.title('Test Accuracy on N24News', fontsize=15, pad=10)

# Customize legend
plt.legend(fontsize=10, title_fontsize=14, loc='lower right', frameon=True)

plt.xticks([5000, 15000, 25000, 35000, 45000], fontsize=12)
plt.yticks([82.5, 83, 83.5, 84], fontsize=12)

# # Add grid lines for better readability
# plt.grid(True, which='both', linestyle='--', linewidth=0.6)

# Display the plot
plt.tight_layout()
plt.show()
```
